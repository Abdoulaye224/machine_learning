```python
dftrain = df.head(30)
dftrain
```

```python
def prepare_data(df, n_input):
    X, y = [], []
    for i in range(n_input, len(df)):
        # extraire les positions et les configurations précédentes
        pos_inputs = df[i-n_input:i, :3].reshape(n_input*3)
        conf_inputs = df[i-n_input:i, 3:5].reshape(n_input*2)
        # extraire la position actuelle
        X.append(np.concatenate((pos_inputs, conf_inputs)))
        y.append(df[i, :3])
    X = np.array(X)
    y = np.array(y)
    return X, y
```

```python
# définir le nombre de positions précédentes à inclure
n_input = 30
```

```python
# préparer les données d'entraînement
train_data = dftrain[['x', 'y', 'z']].values  # sélectionner les colonnes x, y et z et convertir en un tableau numpy
X_train, y_train = prepare_data(train_data, n_input)
```

```python
# préparer les données de test
test_data = dftrain[['x', 'y', 'z']].values  # sélectionner les colonnes x, y et z et convertir en un tableau numpy
X_test, y_test = prepare_data(test_data, n_input)
```

```python
model = SARIMAX(y_train, order=(5, 1, 0))
model_fit = model.fit(method='bfgs', maxiter=1000)

# Print the summary of the fitted model
print(model_fit.summary())

# Generate predictions using the fitted model
y_pred = model_fit.forecast(steps=10)

# Print the predicted values
print(y_pred)
```

```python
model = ARIMA(y_train, order=(5,1,0))
model_fit = model.fit()
# faire des prédictions sur les données de test
y_pred = model_fit.predict(start=len(y_train), end=len(y_train)+len(y_test)-1, typ='levels')
```
