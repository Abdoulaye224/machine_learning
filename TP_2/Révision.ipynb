### Load dataset

```python
donnee = pd.read_csv('dataset.csv')
```

### Data Cleaning

```python
#Remplacer les valeurs vides par 0 pour les colonnes de type entier ou floar
donnee[['InvoiceNo', 'Quantity', 'UnitPrice', 'CustomerID']]=donnee[['InvoiceNo', 'Quantity', 'UnitPrice', 'CustomerID']].fillna(0)
```

```python
donnee['InvoiceDate'] = pd.to_datetime(donnee['InvoiceDate'], errors='coerce')
```

```python
donnee['InvoiceDate']= donnee['InvoiceDate'].dt.strftime("%Y-%m-%d")
```

```python
donnee['CustomerID']= donnee['CustomerID'].astype(int)
```

```python
donnee['UnitPrice']=donnee['UnitPrice'].apply(lambda x: x.replace(',', '.'))
```

```python
donnee['UnitPrice'] = donnee['UnitPrice'].astype(float)
```

```python
donnee['Quantity'] = donnee['Quantity'].abs()
```

```python
donnee['year'] = pd.DatetimeIndex(donnee['InvoiceDate']).year
donnee['month'] = pd.DatetimeIndex(donnee['InvoiceDate']).month
donnee['day'] = pd.DatetimeIndex(donnee['InvoiceDate']).day
```

```python
donnee.head()
```

```python
donnee = donnee.drop_duplicates()
```
